```python
import numpy as np                  # imports the numeric python library
from matplotlib import pyplot as plt   # imports the required libraries for plotting
w = np.arange(1,100,1)                  # vector

x = np.abs(np.exp(2-1j*w)/(w*1j+2))

plt.figure(1)
plt.plot(1*w, x)
plt.xscale('log')
plt.yscale('log') 
plt.xlabel('t')
plt.ylabel('x(t)')
plt.grid()

x = np.abs((np.exp(-2)/(2-1j*w))+(np.exp(2)/(2+1j*w)))

plt.figure(2)
plt.plot(1*w, x)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.grid()
```
