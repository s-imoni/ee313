```python
import numpy as np                  # imports the numeric python library
from matplotlib import pyplot as plt   # imports the required libraries for plotting
#k = np.arange(1,5,1)                  # vector
k = np.array([-4,-3,-2,-1, 1,2,3,4])

x = np.real( (1/(1j*np.pi*k)) * (-.5*((1j)**k) + 1 -.5*((-1.0)**k)) )
h = np.imag( (1/(1j*np.pi*k)) * (-.5*((1j)**k) + 1 -.5*((-1.0)**k)) )

fig, axes = plt.subplots(2, 1, figsize = (6,9))
plt.xlabel("time (sec)")
axes[0].set_title("Real Part of A")
axes[0].stem(k, x, use_line_collection=True)
axes[0].plot(0,.75, "bo")
axes[1].set_title("Imaginary Part of A")
axes[1].stem(k, h, use_line_collection=True)
axes[1].plot(0,0, "bo")
plt.show() 


x = np.real((1/(1j * k * np.pi))*(np.cos(np.pi*2/3.0*k)) - (1/(1j * k * np.pi))*(np.cos(np.pi*1/3.0*k)))
h = np.imag((1/(1j * k * np.pi))*(np.cos(np.pi*2/3.0*k)) - (1/(1j * k * np.pi))*(np.cos(np.pi*1/3.0*k)))

fig, axes = plt.subplots(2, 1, figsize = (6,9))
plt.xlabel("time (sec)")
axes[0].set_title("Real Part of B")
axes[0].stem(k, x, use_line_collection=True)
axes[0].plot(0,0, "bo")
axes[1].set_title("Imaginary Part of B")
axes[1].stem(k, h, use_line_collection=True)
axes[1].plot(0,0, "bo")
axes[1].plot(0,0)
plt.show() 
```
