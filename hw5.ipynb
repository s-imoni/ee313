```python
import numpy as np                  # imports the numeric python library
from matplotlib import pyplot as plt   # imports the required libraries for plotting
t = np.arange(0,10,1)                  # vector

x = (0.7**t)*(np.cos(np.pi*t/4.0)) 
h = (0.8**t)*(np.sin(np.pi*t/3.0))

fig, axes = plt.subplots(2, 1, figsize = (6,9))
plt.xlabel("time (sec)")
axes[0].set_title("x[n]")
axes[0].stem(t, x, use_line_collection=True)
axes[1].set_title("h[n]")
axes[1].stem(t, h, use_line_collection=True)   
plt.show() 

ly = len(x)+len(h)-1;
ty = np.arange(0, ly, 1);
# init empty w 0s
y = np.zeros(ly);
for(i, v) in enumerate(x):
    y[i:i+len(h)] += v*h
    
plt.stem(ty, y, use_line_collection=True)
plt.title("y[n]")
plt.xlabel("n");
plt.ylabel("y[n]");
plt.show()
```
