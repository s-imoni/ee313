## Preliminaries

```python
import numpy as np
import matplotlib.pyplot as plt
np.set_printoptions(precision=3)
```

## Function we use for this problem

We are making use of the Heaviside function found in the newest version of numpy. Heaviside is another name for the unit-step function. 

https://numpy.org/doc/stable/reference/generated/numpy.heaviside.html

```python
def awesome_x(t_as_in_time):
    """ returns value of x(t) at the indicate time of t_as_in_time """
    
    return (1+t_as_in_time)*(np.heaviside(t_as_in_time+1,0.5) - np.heaviside(t_as_in_time,0.5)) \
            + 2*np.heaviside(t_as_in_time,0.5)\
            - np.heaviside(t_as_in_time-1,0.5) \
            - np.heaviside(t_as_in_time-2,0.5)

def awesome_two(t_as_in_time):
    """ returns value of x(t) at the indicate time of t_as_in_time """
    return -(1+t_as_in_time)*(np.heaviside(t_as_in_time+1,0.5) - np.heaviside(t_as_in_time,0.5)) \
            + 0.5*np.heaviside(t_as_in_time,0.5)\
            + 1.5*np.heaviside(t_as_in_time-1,0.5) \
            - 2*np.heaviside(t_as_in_time-2,0.5)  

def awesome_three(t_as_in_time):
    """ returns value of x(t) at the indicate time of t_as_in_time """
    #z(t) = 0.5u(t−2)+u(t)−0.5u(t−4)
    return  .5*np.heaviside(t_as_in_time-2,0.5) + np.heaviside(t_as_in_time,0.5) - 0.5*np.heaviside(t_as_in_time-4,0.5)
            
    
```

## Plotting the function in question

```python
t = np.linspace(-5,5,100) 
x = awesome_two(t)
y = (awesome_two(t)+awesome_two(-t))/2
z = awesome_three(t)
plt.figure(1)
plt.plot(1*t, x)
plt.xlabel('t')
plt.ylabel('x(t)')
plt.grid()
plt.figure(2)
plt.plot(-1*t, x)
plt.xlabel('t')
plt.ylabel('x(t)')
plt.grid()
plt.figure(3)
plt.plot(-2*t, x)
plt.xlabel('t')
plt.ylabel('x(t)')
plt.grid()
plt.figure(4)
plt.plot(-2*t+1, x)
plt.xlabel('t')
plt.ylabel('x(t)')
plt.grid()
plt.figure(5)
plt.plot(.5*t, x)
plt.xlabel('t')
plt.ylabel('x(t)')
plt.grid()
plt.figure(6)
plt.plot(t, y);
plt.xlabel('t')
plt.ylabel('x(t)')
plt.grid()
plt.figure(7)
plt.plot(t, z);
plt.xlabel('t')
plt.ylabel('x(t)')
plt.grid()
plt.show()
```
